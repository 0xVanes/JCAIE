Masukan 5 film kesukaan anda yang dipisahkan dengan koma: Avengers, Hulk, Ironman 3, Wonder woman, Batman
Masukan 5 film kesukaan teman anda dipisahkan dengan koma: Avengers, Hulk, Iron man 2, Superman, Wonder Woman
Kesukaan Film kalian yang sama sebesar 60.0%

```python
# Cara 1
film_Saya = {"Avengers", "Hulk", "Ironman 3", "Wonder woman", "Batman"}
film_teman = {"Avengers", "Hulk", "Ironman 2", "Superman", "Wonder woman"}

film_Saya.update(["Thor", "Loki", "Antman", "Superman", "Ironman 2"])
film_teman.add("Antman")
Kesamaan = film_Saya.intersection(film_teman)
Total_Film = film_Saya.union(film_teman)

Film_Yang_Disukai_Bersama = len(Kesamaan)/len(Total_Film)*100
print(f"{Film_Yang_Disukai_Bersama}%")
```

```python
# Cara 2
film_Saya = {"Avengers", "Hulk", "Ironman 3", "Wonder woman", "Batman"}
film_teman = {"Avengers", "Hulk", "Ironman 2", "Superman", "Wonder woman"}

if film_Saya.issubset(film_teman) == False:
   a = film_Saya.difference(film_teman)
   film_Saya.difference_update(a) #Hapus yang ada di a dan film_Saya

Kesamaan = film_Saya.intersection(film_teman)
Total_Film = film_Saya.union(film_teman)
print(f"{len(Kesamaan)/len(Total_Film)*100}%")
```

```python
# Cara 3
film_Saya = {"Avengers", "Hulk", "Ironman 3", "Wonder woman", "Batman"}
film_teman = {"Avengers", "Hulk", "Ironman 2", "Superman", "Wonder woman"}

a = film_Saya.difference(film_teman)
for film in a:
   film_Saya.discard(film)
   
Kesamaan = film_Saya.intersection(film_teman)
Total_Film = film_Saya.union(film_teman)
print(f"{len(Kesamaan)/len(Total_Film)*100}%")
```

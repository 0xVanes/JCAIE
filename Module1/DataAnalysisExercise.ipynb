```python
import math
```

```python
#KALAU MAU  JADI 2D OTOMATIS GIMANA?
lennyaArray = len(newarray)
# row bisa di square root tapi kalau angkanya kayak 5 atau 7 gimana?
row2d = int(math.sqrt(lennyaArray))
while lennyaArray % row2d != 0:
    row2d -= 1
# Cari dulu columnya berapa
column2d = lennyaArray // row2d
#reshape
anothernewarray2d = newarray.reshape(row2d, column2d)
anothernewarray2d.shape
```

```python
#KALAU MAU JADI 3D OTOMATIS GIMANA?
lennyaArray = len(newarray)
row3d = int(lennyaArray**(1/3))
while lennyaArray % row3d != 0:
    row3d -= 1
#SISA 2 SHAPE
sisa_isi = int(math.sqrt(lennyaArray // row3d))
while sisa_isi % sisa_isi != 0:
    sisa_isi -= 1
#SISA 1 SHAPE
column3d = (lennyaArray//row3d) // sisa_isi
anothernewarray3d = newarray.reshape(row3d, sisa_isi, column3d)
anothernewarray3d.shape
```

```python
# CARA 1 PYTHON
newSeries = pd.Series(data = np.random.randint(1,5,100))
number1 = 0
number2 = 0
number3 = 0
number4 = 0
for i in newSeries:
    if i == 1:
        number1 += 1
    elif i == 2:
        number2 += 1
    elif i == 3:
        number3 += 1
    else:
        number4 += 1
list1 = [number1, number2, number3, number4]
print(f'''1 = {number1}
2 = {number2}
3 = {number3}
4 = {number4}''')
print(list1)

# Cari 2 number dengan frekuensi tertinggi
# CARA ini eror kalau index 0 dan index 1 highest frequency (mungkin lebih bagus di remove, masukin ke list lain)
firsthighest = max(list1)
firstvalue = list1.index(firsthighest) + 1

copylist1 = list1.copy()
copylist1.remove(firsthighest)
secondhighest = max(copylist1)
secondvalue = list1.index(secondhighest) + 1

print(f"The two highest are frequency {firsthighest} and {secondhighest} with value {firstvalue} and {secondvalue}")
```

```python
#CARA 3 PYTHON + SORT
twohighest = sorted(
    newSeries.value_counts().items(), key=lambda x: x[1], reverse=True)[:2]

for value, frequency in twohighest:
    print(f"The two highest are {frequency} with value {value}")
```

```python
#CARA 3 PYTHON + DICT
# Cari 2 number dengan frekuensi tertinggi
dict1 = {1: number1, 2: number2, 3:number3, 4:number4}
highestoftwo = sorted(dict1.items(), key=lambda x: x[1], reverse=True)[:2]
print(f"""The two highest are frequency {highestoftwo[0][1]} and {highestoftwo[1][1]}
    with values {highestoftwo[0][0]} and {highestoftwo[1][0]}""")
```

```python
# CARA 4 NUMPY AND PANDAS
newSeries = pd.Series(data = np.random.randint(1,5,100))
x, y = np.unique(newSeries, return_counts=True)

newCount = pd.DataFrame(data = {"value": x, "frequency": y})
print(newCount)

highesttwo = newCount.sort_values(by="frequency", ascending=False).head(2)
highesttwo
```

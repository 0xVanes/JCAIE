```python
print("Selamat Datang di Pasar Buah")
print('''List Menu:
      1. Menampilkan Daftar Buah
      2. Menambahkan Buah
      3. Menghapus Buah
      4. Membeli Buah
      5. Exit Program''')

#Ubah list jadi dict
Program = {
    0 : {"Nama":"apel", "Stock": 20, "Harga": 10000},
    1 : {"Nama":"jeruk", "Stock": 15, "Harga": 15000},
    2 : {"Nama":"anggur", "Stock": 25, "Harga": 20000},
    }

while True:
    Jwbn = int(input("Masukan angka Menu yang ingin dijalankan: "))
    if Jwbn == 1:
        print(Program)

        print('''List Menu:
            1. Menampilkan Daftar Buah
            2. Menambahkan Buah
            3. Menghapus Buah
            4. Membeli Buah
            5. Exit Program''')

    elif Jwbn == 2:
        search_namabuah = input("Masukkan Nama Buah: ").lower()
        for value_info in Program.values(): #value_info adalah semua value di Program (kalau mau value di dlm dictionary nya lagi tulis Program["Nama"].values())
            if value_info["Nama"] == search_namabuah:
                #value_info itu valuenya 0,1&2
                add_buah = int(input("Masukkan Stock Buah: "))
                value_info["Stock"] += add_buah
                ganti_harga = int(input("Masukkan Harga Buah Baru: "))
                value_info["Harga"] = ganti_harga

                break

        else:
            index = len(Program)
            Program[index] = {}
            Program[index]["Nama"] = search_namabuah
            add_buah = int(input("Masukkan Stock Buah: "))
            Program[index]["Stock"] = add_buah
            harga_buah = int(input("Masukkan Harga Buah: "))
            Program[index]["Harga"] = harga_buah

        print(Program)
        print('''List Menu:
            1. Menampilkan Daftar Buah
            2. Menambahkan Buah
            3. Menghapus Buah
            4. Membeli Buah
            5. Exit Program''')

    elif Jwbn == 3: ##OUTPUT IS TRUNCATED
        print(Program)

        hapus_index = int(input("Masukkan Index yang ingin dihapus = "))
        # IF PROGRAM punya value hapus_index
        if hapus_index in Program:
            del Program[hapus_index]
        else:
            print("Index tidak ada")
        
        print(Program)
        print('''List Menu:
            1. Menampilkan Daftar Buah
            2. Menambahkan Buah
            3. Menghapus Buah
            4. Membeli Buah
            5. Exit Program''')

    elif Jwbn == 4:

        print(Program)
        cart = dict()
        
        while True:
            beli_index = int(input("Masukkan index buah yang ingin dibeli: "))
            
            if beli_index in Program:
                beli_jumlah = int(input("Masukkan jumlah yang ingin dibeli: "))
                if beli_jumlah > Program[beli_index]["Stock"]:
                    print(f"Stock tidak cukup. Stock {Program[beli_index]['Nama']} tinggal {Program[beli_index]['Stock']}")
                    continue
                
                # KURANGIN DARI STOCK
                Program[beli_index]["Stock"] -= beli_jumlah

                # TAMBAH KE CART
                index_baru = len(cart)+1
                cart[index_baru] = {}
                cart[index_baru]["Nama"] = Program[beli_index]["Nama"]
                cart[index_baru]["Jumlah"] = beli_jumlah
                cart[index_baru]["Harga"] = Program[beli_index]["Harga"]

                print(cart)

                buy_or_nah = input("Mau beli yang lain? (ya/tidak) = ").lower()

                if buy_or_nah == "tidak":
                    break
            
            else:
                print("Index buah tidak tersedia")
                
        # CEK TOTAL
        total = 0
        for total_totalin in cart.values():
            total += total_totalin["Jumlah"]*total_totalin["Harga"]


        # PEMBAYARAN
        print(f"Total yang harus dibayar = {total}")
        cash = int(input("Masukkan Jumlah Uang = Rp. "))
        while cash < total:
            print(f"Uang Anda Kurang: Rp. {total-cash}")
            cash = int(input("Masukkan Jumlah Uang = Rp. "))
            continue
        if cash > total:
            print(f"Terima Kasih. Kembalian anda: Rp. {cash-total}")
            break
        else:
            print("Terima Kasih")
            break

    elif Jwbn == 5:
        print("Terima Kasih. Kami Tunggu Kehadiran Anda Kembali")
        break
    else:
        print("Menu tidak tersedia")

```

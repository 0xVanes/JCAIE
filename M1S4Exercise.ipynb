# SOAL 2

```python
# Cuman kepikiran 1 cara
massa = int(input("Masukan berat badan (Kg): "))
tinggi = int(input("Masukan tinggi badan (cm): "))
IMT = massa/((tinggi/100)**2)
if IMT < 18.5:
    print(f"IMT = {IMT}, Berat Badan Kurang")
elif 24.9 >= IMT >= 18.5:
    print(f"IMT = {IMT},Berat Badan Ideal")
elif 29.9 >= IMT >= 25:
    print(f"IMT = {IMT}, Berat Badan Berlebih")
elif 39.9 >= IMT >= 30:
    print(f"IMT = {IMT}, Berat Badan Sangat Berlebih")
else:
    print(f"IMT = {IMT}, Obesitas")
```

# SOAL 3

```python
# CARA 1
Jumlah_Apel = int(input("Jumlah Pembelian Apel = "))
Jumlah_Jeruk = int(input("Jumlah Pembelian Jeruk = "))
Jumlah_Anggur = int(input("Jumlah Pembelian Anggur = "))

Apel = Jumlah_Apel*10000
Jeruk = Jumlah_Jeruk*15000
Anggur = Jumlah_Anggur*20000
Total = Apel+Jeruk+Anggur

print(f"Total Pembelanjaan = {Total}")
Pembayaran = int(input("Masukan Jumlah Uang = "))
if Pembayaran == Total:
    print("Terima Kasih")
elif Pembayaran >= Total:
    a = Pembayaran - Total
    print(f"""Terima Kasih!
          Uang Kembalian anda: {a}""")
else:
    b = Total - Pembayaran
    print(f"""Transaksi Anda Dibatalkan.
          Uang anda kurang sebesar : {b}""")
```

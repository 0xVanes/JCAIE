```python
# CARA 1
Numbers = [41, 5, 1, 3, 89, 32]
Numbers.sort()
Minimum_value = Numbers[0]
print(Minimum_value)
Maximum_value = Numbers[-1]
print(Maximum_value)
```

```python
# CARA 2
Numbers = [41, 5, 1, 3, 89, 32]

Minimum_value = min(Numbers)
Maximum_value = max(Numbers)

print(Minimum_value)
print(Maximum_value)
```

```python
# CARA 3
Numbers = [41, 5, 1, 3, 89, 32]
Minimum_value = Numbers[0]
Maximum_value = Numbers[0]

for n in Numbers:
    if n < Minimum_value:
        Minimum_value = n
    if n > Maximum_value:
        Maximum_value = n

print( Minimum_value, Maximum_value)
```

```python
# CARA 4
Numbers = [41, 5, 1, 3, 89, 32]
Minimum_value = Numbers[0]
Maximum_value = Numbers[0]

for n in range(len(Numbers)):
    if Numbers[n] < Minimum_value:
        Minimum_value = Numbers[n]
    if Numbers[n] > Maximum_value:
        Maximum_value = Numbers[n]

print( Minimum_value, Maximum_value)
```

```python
print("Selamat Datang di Pasar Buah")
print('''List Menu:
      1. Menampilkan Daftar Buah
      2. Menambahkan Buah
      3. Menghapus Buah
      4. Membeli Buah
      5. Exit Program''')

Index = [0,1,2]
Nama = ["apel", "jeruk", "anggur"]
Stock = [20,15,25]
Harga = [10000,15000,20000]

while True:
    Jwbn = int(input("Masukan angka Menu yang ingin dijalankan: "))
    if Jwbn == 1:
        print(Index, Nama, Stock, Harga)

        print('''List Menu:
            1. Menampilkan Daftar Buah
            2. Menambahkan Buah
            3. Menghapus Buah
            4. Membeli Buah
            5. Exit Program''')

    elif Jwbn == 2:
        search_namabuah = input("Masukkan Nama Buah: ").lower()
        if search_namabuah in Nama:
            idx = Nama.index(search_namabuah)
            add_buah = int(input("Masukkan Stock Buah: "))
            Stock[idx] += add_buah
            ganti_harga = int(input("Masukkan Harga Buah Baru: "))
            Harga[idx] = ganti_harga
            
            print(Index, Nama, Stock, Harga)
            print('''List Menu:
                1. Menampilkan Daftar Buah
                2. Menambahkan Buah
                3. Menghapus Buah
                4. Membeli Buah
                5. Exit Program''')

        else:
            Nama.append(search_namabuah)
            add_buah = int(input("Masukkan Stock Buah: "))
            Stock.append(add_buah)
            harga_buah = int(input("Masukkan Harga Buah: "))
            Harga.append(harga_buah)
            Index.append(len(Index))

            print(Index, Nama, Stock, Harga)
            print('''List Menu:
                1. Menampilkan Daftar Buah
                2. Menambahkan Buah
                3. Menghapus Buah
                4. Membeli Buah
                5. Exit Program''')

    elif Jwbn == 3:
        print(Index, Nama, Stock, Harga)

        hapus_index = int(input("Masukkan Index yang ingin dihapus = "))
        del Index[hapus_index]
        del Nama[hapus_index]
        del Stock[hapus_index]
        del Harga[hapus_index]
        print(Index, Nama, Stock, Harga)

        print('''List Menu:
            1. Menampilkan Daftar Buah
            2. Menambahkan Buah
            3. Menghapus Buah
            4. Membeli Buah
            5. Exit Program''')

    elif Jwbn == 4:

        print(Index, Nama, Stock, Harga)
        cart_nama = []
        cart_qty = []
        cart_harga = []
        
        while True:
            beli_index = int(input("Masukkan index buah yang ingin dibeli: "))
            beli_jumlah = int(input("Masukkan jumlah yang ingin dibeli: "))
            
            if beli_jumlah > Stock[beli_index]:
                print(f"Stock tidak cukup. Stock {Nama[beli_index]} tinggal {Stock[beli_index]}")
                continue
            
            # KURANGIN DARI STOCK
            Stock[beli_index] -= beli_jumlah

            # TAMBAH KE CART
            cart_nama.append(Nama[beli_index])
            cart_qty.append(beli_jumlah)
            cart_harga.append(Harga[beli_index]*beli_jumlah)

            print(cart_nama, cart_qty, cart_harga)

            buy_or_nah = input("Mau beli yang lain? (ya/tidak) = ").lower()

            if buy_or_nah == "tidak":
                break
                
        # PEMBAYARAN
        total = sum(cart_harga)
        print(f"Total yang harus dibayar = {total}")
        cash = int(input("Masukkan Jumlah Uang = Rp. "))
        while cash < total:
            print(f"Uang Anda Kurang: Rp. {total-cash}")
            cash = int(input("Masukkan Jumlah Uang = Rp. "))
            continue
        if cash > total:
            print(f"Terima Kasih. Kembalian anda: Rp. {cash-total}")
            break
        else:
            print("Terima Kasih")
            break

    elif Jwbn == 5:
        print("Terima Kasih. Kami Tunggu Kehadiran Anda Kembali")
        break
    else:
        print("Menu tidak tersedia")

```

```python
# PROSES BELI
Stok_Apel = 10
Stok_Jeruk = 7
Stok_Anggur = 6

Jumlah_Apel = int(input("Masukkan jumlah apel = "))
while Jumlah_Apel > Stok_Apel:
    print(f"Stock Apel Tinggal {Stok_Apel}. Jumlah yang dimasukkan terlalu banyak.")
    Jumlah_Apel = int(input("Masukkan jumlah apel = "))

Jumlah_Jeruk = int(input("Masukkan jumlah jeruk = "))
while Jumlah_Jeruk > Stok_Jeruk:
    print(f"Stock Jeruk Tinggal {Stok_Jeruk}. Jumlah yang dimasukkan terlalu banyak.")
    Jumlah_Jeruk = int(input("Masukkan jumlah jeruk = "))

Jumlah_Anggur = int(input("Masukkan jumlah anggur = "))
while Jumlah_Anggur > Stok_Anggur:
    print(f"Stock Anggur Tinggal {Stok_Anggur}. Jumlah yang dimasukkan terlalu banyak.")
    Jumlah_Anggur = int(input("Masukkan jumlah anggur = "))

# DETAIL BELANJA
Apel = Jumlah_Apel*10000
print(f"Apel = {Jumlah_Apel} x 10000 = Rp. {Apel}")
Jeruk = Jumlah_Jeruk*15000
print(f"Jeruk = {Jumlah_Jeruk} x 15000 = Rp. {Jeruk}")
Anggur = Jumlah_Anggur*20000
print(f"Anggur = {Jumlah_Anggur} x 120000 = Rp. {Anggur}")
Total = Apel+Jeruk+Anggur
print(f"Total = Rp. {Total}")

# PROSES BAYAR
Pembayaran = int(input("Masukkan Jumlah Uang = Rp. "))
while True:
    if Pembayaran < Total:
        print(f"Uang and kurang sebesar Rp. {Total-Pembayaran}")
        Pembayaran = int(input("Masukkan Jumlah Uang = Rp. "))
    elif Pembayaran == Total:
        print("Terima Kasih!")
        break
    else:
        print(f"Terima Kasih! Kembalian anda sebesar Rp. {Pembayaran-Total}")
        break
```

SOAL 5

Jarak A-B = 120km
Speed A = 60km/h
Speeb B = 40km/h
Jam 9 pagi A ada di timur, B ada di barat
A sampai ke barat = 120/60 = 2h
B sampai ke timur = 120/40 = 3h
Jam berapa A & B bertabrakan

```python
# CARA 1
h = 120/100
tabrakan1 = 9 + int(h) + (h%1*60*0.01)
print(f"Tabrakan terjadi pada jam {tabrakan1} WIB")

# CARA 2
for x in range(1,120):
    h = x/60
    if 60*h + 40*h == 120: 
        tabrakan2 = 9 + int(h) + h%1*0.6
        print(f"Tabrakan pada jam {tabrakan2} WIB")

# CARA 3
distance = 120
for time_in_minutes in range(1,200):
    distance1 = 60*time_in_minutes/60
    distance2 = 40*time_in_minutes/60
    if distance1 + distance2 == distance:
        jam_tabrakan = 9 + time_in_minutes // 60 + time_in_minutes%60*0.01
        print(f"Tabrakan terjadi pada {jam_tabrakan} WIB")
```
